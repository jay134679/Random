### 3.3.1

```python

def VisualizeTree(X_train, X_test, y_train, y_test, tree_depth, n_classes = 2):
    """  visuualizes the tree split for a data set with n instances and 2 features
    X_train: nx2 2Darray of training data
    X_test: nx2 2Darray of testing data
    y_train: array of length n of training target variable
    y_test: array of length n of training target variable
    tree_depth : int. maximum tree depth
    n_classes: number of classes in target vairable 2,
    """
    # Parameters
    n_classes = 2
    plot_colors = "by"
    plot_step = 0.02
    clf = DecisionTreeClassifier(max_depth=tree_depth).fit(X_train, y_train)

    #create meshgrid
    x_min, x_max = X_test[:, 0].min() - 1, X_test[:, 0].max() + 1
    y_min, y_max = X_test[:, 1].min() - 1, X_test[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                         np.arange(y_min, y_max, plot_step))
    
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    cs = plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha = 0.5)
    
    plt.xlabel("feature 1")
    plt.ylabel("feature 2")

    # Plot the training points
    for i, color in zip([-1,1], plot_colors):
        idx = np.where(y_test == i)
        plt.scatter(X_test[idx, 0], X_test[idx, 1], c=color,
                    cmap=plt.cm.Paired, alpha = 0.2)

        plt.title("Decision surface of a decision tree of depth %s" %tree_depth)
    plt.savefig("Decision surface of a decision tree of depth %s.pdf" %tree_depth, format = 'pdf')
    plt.show()
```

```python
# example of visualization.
VisualizeTree(X_train, X_test, y_train, y_test, tree_depth=4)
```

### 3.3.3

```python
def ErrorByDepth(X_train, X_test, y_train, y_test, max_tree_depth):
    """ This function plots the test and training error as a funciton of the tree depth. """
    
    errors_train = []
    errors_test = []
    
    for tree_depth in range(1,max_tree_depth+1):
        clf = DecisionTreeClassifier(max_depth=tree_depth).fit(X_train, y_train)
        error_fraction_train = 1-clf.score(X_train, y_train)
        error_fraction_test = 1-clf.score(X_test, y_test)
        errors_train.append(error_fraction_train)
        errors_test.append(error_fraction_test)
    
    plt.plot(range(1,max_tree_depth+1), errors_train, label = "train ")
    plt.plot(range(1,max_tree_depth+1), errors_test,label = "test")  
    plt.title("Error by Tree Depth")
    plt.xlabel("Tree depth")
    plt.ylabel("Fraction of Error")
    plt.legend()
    plt.show()
    
    
    
```

```python
ErrorByDepth(X_train, X_test, y_train, y_test, 10)
```

```python
## Analysis:
# In the graph of the variation of error by tree depth, we see several important features:
# a) the training and test error decrease rapidly, dispite being composed of weak learners.
# b) the training error tends towards 0 very quickly.
# c) At a relatively shallow tree depth, we quickly start to overfit the training data. As we can see when tree_depth >6, the 
#    test error becomes worse instead of better.
```

## 5 AdaBoost

### 5.1.1

```python

def AdaBoost(X_train, X_test, y_train, y_test, iterations=1):
    """ Performs the AdaBoost algorithms.
        Args: 
            X_train: nx2 2Darray of training data
            X_test: nx2 2Darray of testing data
            y_train: array of length n of training target variable
            y_test: array of length n of training target variable
            tree_depth : int. maximum tree depth
            n_classes: number of classes in target vairable 2
    
        Returns:
            G: a list of the classifiers we produced with AdaBoost
            weights_cum: a list of list. each sublist contains the weights associated with that round of the AdaBoost algorithm.
            predictions: a list of perdiction for each instance in the testing dataset. {-1,1}.
    
    """
    G = []
    alphas = np.zeros(iterations)
    weights = np.ones(y_train.shape)
    weights_cum = []
    for m in range(iterations):
        clf = DecisionTreeClassifier(max_depth=3)
        clf.fit(X_train, y_train,sample_weight=weights)
        
        #save the classifier for later use
        G.append(clf)
        
        #calculate alpha
        weighted_error = 1-clf.score(X_train, y_train, sample_weight=weights)
        alphas[m] = np.log((1-weighted_error)/weighted_error)
        
        #find misclassifications
        y_predict = clf.predict(X_train)
        diff = y_predict - y_train
        misclassified = [1 if i!=0 else 0 for i in diff]
        
        #assign new weights
        new_weights = [weights[i]*misclassified[i]*np.exp(alphas[m]) if misclassified[i] == 1 else weights[i] for i in range(len(weights))]
        weights_cum.append(weights)
        weights = new_weights

    #calculate predictions for each point in test
    predictions = []
    for idx, point in enumerate(y_test):
        point_prediction = []
        for m,classifier in enumerate(G):
            G_predict = classifier.predict(X_test[idx,:])[0]
            point_prediction.append(alphas[m]*G_predict)
        p = np.sign(sum(point_prediction))
        predictions.append(p)
    return G,predictions,weights_cum


G,predictions,weights_cum= AdaBoost(X_train, X_test, y_train, y_test,iterations=10)
```

```python
def plot_train_test(X_train, X_test, y_train, y_test):
    errors_train=[]
    errors_test=[]
    for i in range(1,11):
        G,p,weights = AdaBoost(X_train, X_test, y_train, y_test,iterations=i)
        errors_test.append(1-accuracy_score(y_test, p))
        G,p,weights = AdaBoost(X_train, X_train, y_train, y_train,iterations=i)
        errors_train.append(1-accuracy_score(y_train, p))
    
    plt.plot(range(1,11),errors_train, label = "train")
    plt.plot(range(1,11),errors_test, label = "test")
    plt.legend()
    plt.show()

plot_train_test(X_train, X_test, y_train, y_test)
```

To solve for the desired expression we first start with the loss function:

$l(f(x),y) = \frac{1}{2}(f(x)-y)^2$ 

We then look at a single instance for the loss function and differentiate it to match the definition of $g_{m,i} $

$l(f(x_i),y_i) = \frac{1}{2}(f(x_i)-y_i)^2$

$g_{m,i} = \frac{\partial{l}}{\partial{f(x_i)}} = f(x_i)-y_i$

We then plug in to the expression of $h_m$ 

$h_m = \arg_{h \in F} \min \sum_{i=1}^n(-g_{m,i} -h(x_i))^2 = \arg_{h \in F} \min \sum_{i=1}^n(-f(x_i)+y_i -h(x_i))^2$
